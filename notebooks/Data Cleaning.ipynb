# Data Cleaning: School Distance Learning 

```python
import pandas as pd 
pd.set_option("display.max_columns", 30)
```

```python
df = pd.read_excel('school_distance_learning.xlsx')  
```

```python
df.head() 
```

## Cleaning columns 

```python
# remove () with bangla text 
df.columns = df.columns.str.replace(r"\([^()]*\)","", regex=True)
```

```python
# remove : 
df.columns = df.columns.str.replace(':','')
```

```python
# remove ? at the end 
df.columns = df.columns.str.replace('?','', regex=True)
```

```python
df.columns = df.columns.str.strip() 
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
df.columns
```

# Removes Baangla within Brackets 

```python
def remove_char(s):
    """Function to remove bangla character inside bracket"""
    return s.str.replace(r"\([^()]*\)","", regex=True)
```

```python
# check function
remove_char(df['gender'])
```

```python
# Numerc data 
df_with_num = df[['age',
                  'how_many_days_did_you_have_classes_in_a_week',
                'how_many_classes_did_you_have_per_day']]
df_with_num.head() 
```

```python
df_with_bangla = df[[
    'are_you_willing_to_participate_in_this_survey',
       'gender', 'class',
       'did_you_attend_any_online_class_before_covid-19_pandemic',
       'have_you_attended_online_classes_regularly',
       'what_was_the_duration_of_each_class',
       'what_type_of_device_did_you_use_in_your_online_class',
       'source_of_internet',
       'did_you_feel_any_mental_disturbance_and_physical_problem_during_online_class',
       'what_kind_of_mental_disturbance_did_you_feel_during_online_class',
       'what_kind_of_physical_problem__did_you_face_during_online_class',
       'difficulty_during_online_classes', 'the_worst_part_of_online_classes',
       'the_best_part_of_online_classes', 'exam_preference']]
```

```python
df_without_bangla = df_with_bangla.apply(remove_char, axis=1)
df_without_bangla.head() 
```

```python
# remove minutes written in bangla 
df_without_bangla['what_was_the_duration_of_each_class'] = df_without_bangla['what_was_the_duration_of_each_class'].str.rstrip('মিনিট')
```

```python
# combined data
data = pd.concat([df_with_num, df_without_bangla], axis=1)
data.head() 
```

```python
# check the number of records 
data.shape
```

```python
# data with consent 
data['are_you_willing_to_participate_in_this_survey'].value_counts() 
```

```python
data.to_csv("cleaned_school_distance_learning.csv", index=False)
```
